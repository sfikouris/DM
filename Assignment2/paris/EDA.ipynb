```python
import pandas as pd
```

```python
df = pd.read_csv('training_set_VU_DM.csv/training_set_VU_DM.csv')
```

```python
df.head(50)
```

```python
pd.set_option('display.max_columns',54)
```

```python
df.describe()
```

```python
srch_grp = df.groupby(['srch_id'])
```

```python
AvgHotelPricePerSrchID = srch_grp['price_usd'].median()
AvgHotelPricePerSrchID
```

```python
filt = df['booking_bool'] == 1
df_newbook=df.rename(columns={'price_usd': 'bookedPrice'})
PriceOfBookedValue=df_newbook[filt].set_index('srch_id')['bookedPrice']
```

```python
PriceOfBookedValue
```

```python
filt = df['click_bool'] == 1
df_newclick=df.rename(columns={'price_usd': 'clickedPrice'})
PriceOfclickedValue = df_newclick[filt].set_index('srch_id')['clickedPrice']
PriceOfclickedValue
```

```python
python_df = pd.concat([AvgHotelPricePerSrchID, PriceOfBookedValue, PriceOfclickedValue], axis='columns', sort= False)
python_df
```

```python
python_df.rename(columns={'price_usd':'MedianPricePerQuerry'})
```
